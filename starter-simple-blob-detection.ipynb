```python
#importing packages for image proccessing

import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt
from skimage.feature import blob_dog
```

```python
def Blob(img):
    #starting reading an image and using a blur filter 
    n=3
    img1 = cv.imread(img,cv.IMREAD_COLOR)
    img1 = cv.cvtColor(img1,cv.COLOR_BGR2RGB)
    
    ima = cv.imread(img,cv.IMREAD_GRAYSCALE)
    ima = cv.blur(ima,(n,n))
    
    # apling  threshold for all vaalues that difer from white inverting the image values
    ret, ima = cv.threshold(ima, 254, 255, cv.THRESH_BINARY_INV)
    
    x=blob_dog(ima)
    
    #Center point of the star
    #print('point at ',x)
    
    #adding circles where there where find stars
    for r in range(len(x)):
        c=plt.Circle((x[r][1],x[r][0]),x[r][2],color=(0,0.9,0),linewidth=2, fill=False)
        fig=plt.gcf()
        ax = fig.gca()
        ax.add_patch(c)
    
    #showing the image
    plt.imshow(img1)
    plt.show()
        
    
```

```python
#building a sample of 40 images of the stars

for r in range(1,42):
    Blob('../input/simple-object-detection/datasets/images/a '+'(' +str(r)+ ')' + '.jpg')
```
