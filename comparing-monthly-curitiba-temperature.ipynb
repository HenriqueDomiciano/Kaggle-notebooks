# Starting the Analysis
Our objectives: 
1. Ploting the temperatures of the months across the years at the city of Curitiba
2. Finding correlations between years of variation of temperatures across the months.
3. Is the city of Curitiba getting hotter?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Defining functions to complete missing data, when the year does not have data we supose as it was equal to the previous year.

```python
def normalize(X):
    delta=[0]
    for u in range(1,len(X)) :
        if X[u] > 50:
            X[u]=X[u-1]
        delta.append(X[u]-X[u-1])
    return X,delta
```

Now we are going to build functions to plot the diference of temperature of a year to the previous one and the temperatures across the years.

```python
def plot_months(y,x,legend):
    plt.figure(figsize=(12,6))
    plt.title('Temperature at '+legend)
    plt.plot(y,x,label='temperature')
    plt.legend()
    plt.show()

def plot_delta(y,delta,legend):
    plt.figure(figsize=(12,6))
    plt.title('Variation of average temperature of the year compeare to the previous'+ ' at the month '+legend)
    plt.scatter(y,delta,label='variation year by year')
    plt.xlabel('years')
    plt.legend()
    plt.show()
```

Now we need to read the data using pandas.

```python
df = pd.read_csv('../input/temperature-timeseries-for-some-brazilian-cities/station_curitiba.csv')
df.head()
df.describe()
```

Now we are going to get the data of the year averages and get the years as the y variable

```python
y=df.pop('YEAR')
X=df['metANN']
meses=[]
```

Now we are going to get the monthly values.

```python
colunas=df.columns
for r in range(12):
    meses.append(df[colunas[r]])
meses=np.array(meses)
```

Now we are going to start ploting the year values

```python
X=np.array(X)

X,delta=normalize(X)

plot_months(y,X,'Years')
plot_delta(y,delta,"Years")
```

Here we can see that the data starts to get more sparse comparing the variation year by year. But the temperature is increasing across the years.

Now we will plot the month values.

```python
for r in range(len(meses)):
    plot_months(y,normalize(meses[r])[0],colunas[r])
    plot_delta(y,normalize(meses[r])[1],colunas[r])
```
