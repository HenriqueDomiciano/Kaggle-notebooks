# Seeing the Data and showing correlations 

* First we are goig to analize the relations between age and other variables 
* Later we are going to see the relations between glucose and other variables 
* Last we are going to see the relation between BMI and Insulin
* And then we are going to build a Random Forest algorithm and an MLP algorithm 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import seaborn as sns
```

```python
#reading dataframe

df=pd.read_csv(r'../input/pima-indians-diabetes-database/diabetes.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
corr=df.corr()
sns.heatmap(corr,annot=True)
```

```python
#loking at pregnancies 

plt.title("Pregnancies")
df['Pregnancies'].value_counts().sort_values().plot(kind='barh')
plt.show()
```

```python
#seeing if diabetes afects age

idosos=df.loc[(df['Age']>59)]

a=np.array(idosos['Outcome'].value_counts())
plt.title("Diabetes with people above 60 years of age")
plt.bar(['Does not have diabetes','Have diabetes'],a)
plt.show()

novos=df.loc[(df['Age']<60)]

a=np.array(novos['Outcome'].value_counts())
plt.title("Diabetes with people above under 60 years of age")
plt.bar(['Does not have diabetes','Have diabetes'],a)
plt.show()

a1=np.array(df['Outcome'].value_counts())
plt.title("Diabetes")
plt.bar(['Does not have diabetes','Have diabetes'],a)
plt.show()
```

```python
#analizing average value of glucose data in relation to age

#removing outliers
df=df.loc[(df['Glucose']>10)]

m=df['Age']
k=df['Glucose']

plt.title("Relation between Glucose and Age")
plt.xlabel('Glucose')
plt.ylabel('Age')
plt.scatter(k,m)
plt.show()

plt.title("Histogram Glucose")
plt.ylabel('Number of people')
plt.xlabel("Glucose")
plt.hist(k,10)
plt.show()

#histogram of age
plt.hist(m,10)
plt.ylabel('Number of People')
plt.xlabel("Age")
plt.show()
```

```python
#looking at blood pressure

#removing outliers

df=df.loc[(df['BloodPressure']>0)]

k=df['BloodPressure']
m=df['Age']

plt.title("Relation between Blood Pressure and Age")
plt.ylabel('Age')
plt.xlabel('Blood Pressure')
plt.scatter(k,m)
plt.show()

plt.title("Histogram of Blood Pressure")
plt.ylabel('Number of People')
plt.xlabel('Blood Pressure')
plt.hist(k,10)
plt.show()
```

```python
#looking at 

#removing outliers

df=df.loc[(df['SkinThickness']>0)]

k=df['SkinThickness']
m=df['Age']

plt.title("Relation between Skin Thickness and Age")
plt.ylabel('Age')
plt.xlabel('Skin Thickness')
plt.scatter(k,m)
plt.show()

plt.title("Histogram of Skin Thickness")
plt.ylabel('Number of People')
plt.xlabel('Blood Skin Thickness')
plt.hist(k,10)
plt.show()
```

```python
#looking at insulin

#removing outliers

a=df.loc[(df["Insulin"]>0)]

m=a["Age"]
k=a['Insulin']

plt.title("Relation between Insulin and Age")
plt.scatter(k,m)
plt.xlabel('Insulin')
plt.ylabel('Age')
plt.show()

plt.title("Histogram of Insulin")
plt.ylabel('Number of people')
plt.xlabel('Insulin')
plt.hist(k,10)
plt.show()
```

```python

#looking at BMI


m=df["Age"]
k=df['BMI']

plt.title("Relation between BMI and Age")
plt.scatter(k,m)
plt.show()

plt.title("Histogram of BMI")
plt.ylabel("Number of people")
plt.xlabel("BMI")
plt.hist(k,10)
plt.show()
```
